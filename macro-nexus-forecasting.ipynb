#### Data Preparation

```python
# Load the datasets
inf = pd.read_excel('Inflation rates from upto march.xlsx')
macro = pd.read_csv('macro_data_63_months.txt')

inf['month'] = pd.to_datetime(inf['month'])
macro['month'] = pd.to_datetime(macro['month'])

# Inner merge the two data
df = pd.merge(inf, macro, on= 'month', how = 'inner')

# Data observation
print(df.head())
print(df.info())
print(df.describe())
```
